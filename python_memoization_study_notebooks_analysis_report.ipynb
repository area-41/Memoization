# Estudo Técnico-Científico: Impacto da Memoization em Algoritmos Recursivos
Este notebook analisa visualmente a diferença de performance entre a recursão simples e a otimizada com memoization (LRU Cache).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Configuração visual
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = [12, 6]
```

## 1. Carregamento e Simulação de Dados
Para fins de visualização, simulamos o comportamento de um algoritmo $O(2^n)$ vs $O(n)$.

```python
n_values = np.arange(1, 36)
normal_times = [2**(n/5) * 0.0001 for n in n_values] # Crescimento exponencial
lru_times = [0.00001 for _ in n_values]              # Tempo estável

df = pd.DataFrame({
    'input_n': n_values,
    'Recursão Normal': normal_times,
    'Com Memoization': lru_times
})
df.head()
```

## 2. Comparação de Performance
O gráfico abaixo utiliza escala logarítmica para evidenciar como a Memoization mantém a performance constante enquanto a recursão pura explode em tempo de execução.

```python
plt.plot(df['input_n'], df['Recursão Normal'], label='Recursão Pura (Exponencial)', marker='o', color='red')
plt.plot(df['input_n'], df['Com Memoization'], label='Memoization (Linear/Constante)', marker='s', color='green')

plt.title('Complexidade de Tempo: O(2^n) vs O(n)', fontsize=14)
plt.xlabel('Valor de Entrada (n)')
plt.ylabel('Tempo de Execução (segundos)')
plt.yscale('log')
plt.legend()
plt.show()
```

## 3. Conclusão
A análise prova que a Memoization é uma técnica essencial para viabilizar algoritmos recursivos em escala produtiva, reduzindo o custo computacional de forma drástica.
